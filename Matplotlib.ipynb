```python
#Contour Plots in Python | using Matplotlib
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-2,2,500)
```

```python
y=np.linspace(-2,2,50)
```

```python
X, Y = np.meshgrid(x,y)
```

```python
Z= np.sin(X) + np.cos(Y)
```

```python
plt.ion()
```

```python
plt.figure()
```

```python
plt.contour(X,Y,Z)
plt.colorbar()
plt.clf()
plt.contourf(X,Y,Z)
plt.colorbar()
plt.clf()
plt.contourf(X,Y,Z, levels=30, vmin=-1, vmap=0.5)
```

```python
plt.colorbar()
```

```python
plt.clf()
```

```python
Z = (X**2 + Y**2 -1)**3 - X**2 * Y**3
```

```python
plt.contour(X, Y, Z)
plt.contourf(X, Y, -Z, cmap="Reds")
```
